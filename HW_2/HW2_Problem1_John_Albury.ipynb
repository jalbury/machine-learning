# Homework 2, Problem 1

## Description

In this problem, we are asked to find three simple bivariate functions $f_1$, $f_2$, and $f_3$ and three points $p_1$, $p_2$, and $p_3$ such that:

- $p_1$ is a minimum of $f_1$
- $p_2$ is a maximum of $f_2$
- $p_3$ is neither a minimum nor maximum of $f_3$, but the gradient of $f_3$ at $p_3$ is the zero vector ($p_3$ is a saddle point of $f_3$)

Finding the points where $\nabla f(x, y) =
\begin{pmatrix}
 f_x(x, y) \\
 f_y(x, y)
\end{pmatrix}
 = \bf{0}$ will give us the possible local minima, local maxima, and saddle points. However, this information alone will not tell us which of the three possible categories each of these points falls into. To determine whether a point $(x_0, y_0)$ (where $\nabla f(x_0, y_0) = 0$) is a local minimum, local maximum, or saddle point, the second partial derivative test will be used.

The second partial derivative test (source: [Khan Academy](https://www.khanacademy.org/math/multivariable-calculus/applications-of-multivariable-derivatives/optimizing-multivariable-functions/a/second-partial-derivative-test)) is as follows. First, we compute the following second partial derivatives of $f$ at a point $(x_0, y_0)$ (where $\nabla f(x_0, y_0) = \bf{0}$): $f_{xx}(x_0, y_0)$, $f_{yy}(x_0, y_0)$, and $f_{xy}(x_0, y_0)$. Then, we compute $H = f_{xx}(x_0, y_0)f_{yy}(x_0, y_0) - f_{xy}(x_0, y_0)^2$. Finally, we can classify the point $(x_0, y_0)$ as follows:

- If $H < 0$, then $(x_0, y_0)$ is a saddle point.
- If $H > 0$:
  - If $f_{xx}(x_0, y_0)$ < 0, then $(x_0, y_0)$ is a local maximum.
  - If $f_{xx}(x_0, y_0)$ > 0, then $(x_0, y_0)$ is a local minimum.
- If $H=0$, then we cannot classify whether $(x_0, y_0)$ is a local minimum, local maximum, or saddle point.

For plotting, I used `matplotlib`. The three-dimensional plots are based on code in [Dr. Wocjan's 3D plotting notebook](https://colab.research.google.com/drive/1btvbObh-nZ4MSC7QkjpS3RGpefN_msth#scrollTo=T0ldsRUjv-VF), and the density and contour plots are based on code in [Dr. Wocjan's density and contour plots notebook](https://colab.research.google.com/drive/1pcvtvK6jITbp1Sf2nD2uEaDGpwUOA3IL).

## Solution


### Imports

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting functions

```
# given a bivariate function f, returns 3D points so that f can be plotted
def generate_3d_data(f):
  x = np.linspace(-6, 6, 30)
  y = np.linspace(-6, 6, 30)

  X, Y = np.meshgrid(x, y)
  Z = f(X, Y)

  return (X, Y, Z)

# creates and displays a 3D plot
def plot_3d(surface, surface_label, point, point_label):
  ax = plt.axes(projection='3d')
  ax.plot_surface(surface[0], surface[1], surface[2], rstride=1, cstride=1,
                  cmap='viridis', edgecolor='none', alpha=0.4)
  ax.scatter(point[0], point[1], point[2], color='red', label=point_label);
  ax.legend();
  ax.set_title(surface_label);

# creates and displays a density plot
def plot_density(surface, surface_label, point, point_label):
  plt.contourf(surface[0], surface[1], surface[2], 20, cmap='RdGy')
  plt.plot(point[0], point[1], 'ro', label=point_label);
  plt.legend();
  plt.colorbar();
  plt.title(surface_label);

# creates and displays a contour plot
def plot_contour(surface, surface_label, point, point_label):
  plt.contour(surface[0], surface[1], surface[2], colors='black')
  plt.plot(point[0], point[1], 'ro', label=point_label);
  plt.legend();
  plt.title(surface_label);
```

### Solution for $f_1$ and $p_1$

#### Description

Let $f_1 = x^2 + y^2$. Now, I will apply the second partial derivative test as described in the beginning of this notebook.

$\nabla f(x, y) =
\begin{pmatrix}
 f_x(x, y) \\
 f_y(x, y)
\end{pmatrix} = 
\begin{pmatrix}
 2x \\
 2y
\end{pmatrix}$
 
$\nabla f(x, y) = \bf{0}$ at the point $(0, 0)$

$f_{xx}(x, y) = 2$ and $f_{xx}(0, 0) = 2$

$f_{yy}(x, y) = 2$ and $f_{yy}(0, 0) = 2$

$f_{xy}(x, y) = 0$ and $f_{xy}(0, 0) = 0$

$H = f_{xx}(0, 0)f_{yy}(0, 0) - f_{xy}(0, 0)^2 = 2*2 - 0^2 = 4$

Since $H > 0$ and $f_{xx}(0, 0) > 0$, $(0, 0)$ is a minimum of $f_1 = x^2 + y^2$.

$f_1(0, 0) = 0$. Thus, $p_1 = (0, 0, 0)$.

#### Define $f_1$ and $p_1$

```
def f1(x, y):
    return x ** 2 + y ** 2

surface1 = generate_3d_data(f1)
surface1_label = '$f_1(x, y) = x^2 + y^2$'

point1 = (0, 0, 0)
point1_label = '$p_1 = (0, 0, 0)$'
```

#### Plot $f_1$ and $p_1$

##### 3D Plot

```
plot_3d(surface1, surface1_label, point1, point1_label)
```

##### Density Plot

```
plot_density(surface1, surface1_label, point1, point1_label)
```

##### Contour Plot

```
plot_contour(surface1, surface1_label, point1, point1_label)
```

### Solution for $f_2$ and $p_2$

#### Description

Let $f_2 = -x^4 - x^2 - y^4 - y^2$. Now, I will apply the second partial derivative test as described in the beginning of this notebook.

$\nabla f(x, y) =
 \begin{pmatrix}
  f_x(x, y) \\
  f_y(x, y)
 \end{pmatrix}
= 
 \begin{pmatrix}
  -4x^3 - 2x \\
  -4y^3 - 2y
 \end{pmatrix}$
 
 $\nabla f(x, y) = \bf{0}$ at the point $(0, 0)$

$f_{xx}(x, y) = -12x^2 - 2$ and $f_{xx}(0, 0) = -2$

$f_{yy}(x, y) = -12y^2 - 2$ and $f_{yy}(0, 0) = -2$

$f_{xy}(x, y) = 0$ and $f_{xy}(0, 0) = 0$

$H = f_{xx}(0, 0)f_{yy}(0, 0) - f_{xy}(0, 0)^2 = -2*-2 - 0^2 = 4$

Since $H > 0$ and $f_{xx}(0, 0) < 0$, $(0, 0)$ is a maximum of $f_2 = -x^4 - x^2 - y^4 - y^2$.

$f_2(0, 0) = 0$. Thus, $p_2 = (0, 0, 0)$.

#### Define $f_2$ and $p_2$

```
def f2(x, y):
  return - (x ** 4) - (x ** 2) - (y ** 4) - (y ** 2)

surface2 = generate_3d_data(f2)
surface2_label = '$f_2(x, y) = -x^4 - x^2 - y^4 - y^2$'

point2 = (0, 0, 0)
point2_label = '$p_2 = (0, 0, 0)$'
```

#### Plot $f_2$ and $p_2$

##### 3D Plot

```
plot_3d(surface2, surface2_label, point2, point2_label)
```

##### Density Plot

```
plot_density(surface2, surface2_label, point2, point2_label)
```

##### Contour Plot

```
plot_contour(surface2, surface2_label, point2, point2_label)
```

### Solution for $f_3$ and $p_3$

#### Description

Let $f_3 = x^3 + y^3 + xy$. Now, I will apply the second partial derivative test as described in the beginning of this notebook.

$\nabla f(x, y) =
 \begin{pmatrix}
  f_x(x, y) \\
  f_y(x, y)
 \end{pmatrix}
= 
 \begin{pmatrix}
  3x^2 + y \\
  3y^2 + x
 \end{pmatrix}$
 
 $\nabla f(x, y) = \bf{0}$ at the point $(0, 0)$

$f_{xx}(x, y) = 6x$ and $f_{xx}(0, 0) = 0$

$f_{yy}(x, y) = 6y$ and $f_{yy}(0, 0) = 0$

$f_{xy}(x, y) = 1$ and $f_{xy}(0, 0) = 1$

$H = f_{xx}(0, 0)f_{yy}(0, 0) - f_{xy}(0, 0)^2 = 0*0 - 1^2 = -1$

Since $H < 0$, $(0, 0)$ is a saddle point of $f_3 = x^3 + y^3 + xy$.

$f_3(0, 0) = 0$. Thus, $p_3 = (0, 0, 0)$.

#### Define $f_3$ and $p_3$

```
def f3(x, y):
  return x ** 3 + y ** 3 + x*y

surface3 = generate_3d_data(f3)
surface3_label = '$f_3(x, y) = x^3 + y^3 + xy$'

point3 = (0, 0, 0)
point3_label = '$p_3 = (0, 0, 0)$'
```

#### Plot $f_3$ and $p_3$

##### 3D Plot

```
plot_3d(surface3, surface3_label, point3, point3_label)
```

##### Density Plot

```
plot_density(surface3, surface3_label, point3, point3_label)
```

##### Contour Plot

```
plot_contour(surface3, surface3_label, point3, point3_label)
```
